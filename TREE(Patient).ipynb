```python
X=data.iloc[:,2:12]
Y=data.iloc[:,1] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,train_size=0.8,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_=DecisionTreeClassifier(max_depth=4)
```

```python
tree_patient=tree_.fit(X_train,Y_train)
                      
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
```

```python
fig=plt.figure(figsize=(30,40))
tree.plot_tree(tree_patient,feature_names=list(X.columns.values),class_names=list(Y.values),filled=True)
plt.show()
```

```python
fig.savefig("arbol_paciente.png")
```

```python
Y_pred=tree_patient.predict(X_test)
Y_pred 
```

```python
from sklearn.metrics import confusion_matrix
matrix_confusion=confusion_matrix(Y_test,Y_pred)
matrix_confusion
```

```python
press_tot=np.sum(matrix_confusion.diagonal()) / np.sum(matrix_confusion)
press_tot
```
